```python
from PIL import Image
import csv
import os

import pytesseract
pytesseract.pytesseract.tesseract_cmd = os.path.join('C:\\', 'Program Files', 'Tesseract-OCR', 'tesseract.exe')
```

```python
res = pytesseract.image_to_data(Image.open('C:/Users/Colin/Downloads/azer.png'))
```

```python
reader = csv.DictReader(res.split("\n"), delimiter="\t", quoting=csv.QUOTE_NONE)
for l in reader:
    print(f"level={l['level']}")
    print(f"left={l['left']}, top={l['top']}, width={l['width']}, height={l['height']}")
```
